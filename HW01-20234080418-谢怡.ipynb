已知：
\[
\mathbf{a} = \begin{bmatrix}2 \\ -1 \\ 3\end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix}1 \\ 4 \\ -2\end{bmatrix}
\]
\[
A = \begin{bmatrix}
2 & 1 & -1 \\
-1 & 0 & 2 \\
3 & 1 & 0
\end{bmatrix}, \quad
B = \begin{bmatrix}
1 & 3 & 1 \\
-2 & 0 & 1 \\
0 & 2 & 1
\end{bmatrix}
\]

**1. 向量点积**  
\[
\mathbf{a} \cdot \mathbf{b} = 2\times1 + (-1)\times4 + 3\times(-2) = 2 -4 -6 = -8
\]

**2. 矩阵乘法 \(A \times B\)**  
形状：\(3 \times 3\)。  
元素计算（示例）：
\[
(AB)_{11} = 2\cdot1 + 1\cdot(-2) + (-1)\cdot0 = 0
\]
\[
(AB)_{12} = 2\cdot3 + 1\cdot0 + (-1)\cdot2 = 6 +0 -2 = 4
\]
\[
(AB)_{13} = 2\cdot1 + 1\cdot1 + (-1)\cdot1 = 2+1-1=2
\]
其余元素同理可得。

**3. 向量 \(\mathbf{a}\) 的 Frobenius 范数**  
\[
\|\mathbf{a}\|_F = \sqrt{2^2 + (-1)^2 + 3^2} = \sqrt{4+1+9} = \sqrt{14}
\]
